```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
total_data = {
    "Twitter_tweets": 0,
    "Reddit_comments": 0,
    "Instagram_comments":0,
    "Facebook_comments": 0
}

tweets_data = {}
reddit_data = {}
instagram_data = {
    'Gov.sg':{},
    'Mothership':{},
    'Straits Times':{}
}
```

# Twitter

### Num records per policies

```python
mypath = ".."
folder_name='Data/Twitter Data/Cleaned Data/Policies'
file_path = f'{mypath}/{folder_name}/'
# folders = [folder for folder in os.listdir(file_path) if (folder != '.DS_Store' and folder.endswith('csv') == False)]
files = [filename for filename in os.listdir(file_path) if filename.endswith('.csv')]
files
```

```python
for file in files:
    df = pd.read_csv(f'{file_path}{file}')
    total_data['Twitter_tweets'] += len(df)

    key_word = " ".join(file[:-4].split('_'))
    tweets_data[key_word] = len(df)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0,2,2])
key_words = list(tweets_data.keys())
num_records = list(tweets_data.values())
ax.bar(key_words, num_records, color='royalblue')
plt.title("Number of Twitter records for Covid-19 policies", fontsize = 25, pad=15)
plt.xlabel("Keywords", fontsize = 20)
plt.ylabel("Number of records",fontsize = 20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
for i in range(len(key_words)):
    plt.text(x= i-0.13 , y = num_records[i]+40, s=f"{num_records[i]}" , fontdict=dict(fontsize=15))

plt.tight_layout()
plt.show()
```

# Instagram

```python
mypath = ".."
folder_name_1='Data/Instagram Data/Cleaned Data/Gov.sg'
folder_name_2='Data/Instagram Data/Cleaned Data/Mothership'
folder_name_3='Data/Instagram Data/Cleaned Data/Straits Times'
file_path_1 = f'{mypath}/{folder_name_1}/'
file_path_2 = f'{mypath}/{folder_name_2}/'
file_path_3 = f'{mypath}/{folder_name_3}/'
files_1 = [filename for filename in os.listdir(file_path_1) if filename.endswith('.csv')]
files_2 = [filename for filename in os.listdir(file_path_2) if filename.endswith('.csv')]
files_3 = [filename for filename in os.listdir(file_path_3) if filename.endswith('.csv')]
files_1
```

```python
for file in files_1:
    df = pd.read_csv(f'{file_path_1}{file}')
    total_data['Instagram_comments'] += len(df)

    keyword = file[:-4]
    instagram_data['Gov.sg'][keyword] = len(df)

for file in files_2:
    df = pd.read_csv(f'{file_path_2}{file}')
    total_data['Instagram_comments'] += len(df)

    keyword = file[:-4]
    instagram_data['Mothership'][keyword] = len(df)

for file in files_3:
    df = pd.read_csv(f'{file_path_3}{file}')
    total_data['Instagram_comments'] += len(df)

    keyword = file[:-4]
    instagram_data['Straits Times'][keyword] = len(df)
```

```python
num_comments_per_keyword = {}

for key in instagram_data.keys():
    for k, v in instagram_data[key].items():
        if k not in num_comments_per_keyword:
            num_comments_per_keyword[k] = v
        else:
            num_comments_per_keyword[k] += v
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0,2,2])
key_words = list(num_comments_per_keyword.keys())
num_records = list(num_comments_per_keyword.values())
ax.bar(key_words, num_records, color='darkmagenta')
plt.title("Number of Instagram comments for Covid-19 policies", fontsize = 25, pad=15)
plt.xlabel("Keywords", fontsize = 20)
plt.ylabel("Number of records",fontsize = 20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
for i in range(len(key_words)):
    plt.text(x= i-0.10 , y = num_records[i]+8 , s=f"{num_records[i]}" , fontdict=dict(fontsize=15))

plt.tight_layout()
plt.show()
```

# Merging all data

```python
df = []
count = 0

path = '..'
data = 'Data'
platforms = [folder for folder in os.listdir(f'{path}/Data') if (folder != 'Facebook' and folder != '.DS_Store')]

for platform in platforms:
    folders = [folder for folder in os.listdir(f'{path}/Data/{platform}/Cleaned Data/') if (folder.endswith('csv') == False and folder != '.DS_Store')]

    for folder in folders:
        files = [file for file in os.listdir(f'{path}/Data/{platform}/Cleaned Data/{folder}')]
        for file in files:
            temp = pd.read_csv(f'{path}/Data/{platform}/Cleaned Data/{folder}/{file}')
            count += len(temp)
```
